```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'per capita income (US$)': 'per_capita'})
```

```python
df.columns
```

```python
plt.xlabel('per capita income in usd')
plt.ylabel('year')
plt.scatter(df.per_capita,df.year, color='g')
plt.show()
```

```python
reg=linear_model.LinearRegression()
#fit mein xaxis ka aisa and y as
reg.fit(df[['per_capita']],df.year)
```

```python
reg.predict([[10000]])
```

```python
plt.xlabel('per capita income in usd')
plt.ylabel('year')
plt.scatter(df['per_capita'], df['year'], color='g')
plt.plot(df['per_capita'], reg.predict(df[['per_capita']]), color='blue')
plt.show()
```
